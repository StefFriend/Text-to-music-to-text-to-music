Import section

```python
import ipywidgets as widgets #for the dialog box
from IPython.display import display #for the dialog box
from transformers import AutoProcessor, MusicgenForConditionalGeneration #to generate music
from IPython.display import Audio #to listen on the fly
import scipy #to output wavs
```

Generate prompt box

```python
# Create a text box for input
text_box = widgets.Text(
    value='',
    placeholder='Describe the music YOU want',
    description='Prompt:',
    disabled=False
)

# Display the text box
display(text_box)
```

Generate music

```python
# Load your model and processor
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")

# Retrieve the input from the text box
prompt_text = text_box.value

# Process the input
inputs = processor(text=[prompt_text], padding=True, return_tensors="pt")

# Generate audio
audio_values = model.generate(**inputs, max_new_tokens=128)
```

Listen to generated music

```python

sampling_rate = model.config.audio_encoder.sampling_rate
Audio(audio_values[0].numpy(), rate=sampling_rate)
```

Save output to wav file

```python

sampling_rate = model.config.audio_encoder.sampling_rate

# Format the filename with the prompt text, replacing spaces with underscores and limiting its length
filename = "{}.wav".format(prompt_text.replace(" ", "_")[:50])

scipy.io.wavfile.write(filename, rate=sampling_rate, data=audio_values[0, 0].numpy())
```
